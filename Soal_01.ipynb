Membuat array

```python
nilai = []
```

Membuat Program Input Nilai

```python
for i in range(10):
  n = int(input(f"Masukkan nilai ke-{i+1}: "))
  nilai.append(n)
```

Membuat Program Menghitung Jumlah Lulus

```python
lulus = 0
for n in nilai:
  if n >= 60:
    lulus += 1
```

Membuat Output

```python
print("========HASIL========")
print("Daftar Nilai       : ", nilai)
print("Nilai Tertinggi    : ", max(nilai))
print("Nilai Terendah     : ", min(nilai))
print("Rata-rata          : ", sum(nilai)/len(nilai))
print("Jumlah Lulus       : ", lulus)
print("Jumlah Tidak Lulus : ", len(nilai)-lulus)
```
