Array Nilai

```python
import numpy as np

#Memakai nilai yang tadi sudah diinput di file Soal_01
nilai = np.array([90, 78, 98, 56, 67, 54, 70, 60, 49, 95])
print(("Daftar Nilai : "), nilai)

nilaiTertinggi = max(nilai)
nilaiTerendah = min(nilai)

lulus = 0
for n in nilai:
  if n >= 60:
    lulus += 1

tidakLulus = len(nilai)-lulus
```

Grafik Batang Nilai Tertinggi & Terendah

```python
import matplotlib.pyplot as plt

label = ["Tertinggi", "Terendah"]
data = [nilaiTertinggi, nilaiTerendah]

plt.bar(label, data)
plt.title("Grafik Perbandingan Nilai")
plt.ylabel("Nilai")
plt.show()
```

Grafik Pie Data Kelulusan

```python
import matplotlib.pyplot as plt

label2 = ["Lulus", "Tidak Lulus"]
data2 = [lulus, tidakLulus]

plt.pie(data2, labels=label2, autopct='%1.1f%%')
plt.title("Grafik Pie Chart Persentase Kelulusan")
plt.axis('equal')
plt.show()
```
